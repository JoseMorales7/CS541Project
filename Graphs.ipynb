```python
#base cnn vs base ViT
import numpy as np
import matplotlib.pyplot as plt
```

```python
cnn_train = np.load("./results/inception/Inception_pretrained_train.npy")
cnn_valid = np.load("./results/inception/Inception_pretrained_valid.npy")
cnn_valid_acc = np.load("./results/inception/Inception_pretrained_valid_acc.npy")

vit_train = np.load("./results/vit_b_16/vit_b_16_city4_train.npy")
vit_valid = np.load("./results/vit_b_16/vit_b_16_city4_test.npy")
vit_valid_acc = np.load("./results/vit_b_16/vit_b_16_city4_test_acc.npy")
print(len(cnn_train))
print(cnn_train.shape)
print(len(vit_train))

minLength = min([len(cnn_train), len(vit_train)])
fig, ax = plt.subplots(1)

ax.set_xlabel("Epoch")
ax.set_ylabel("CE Loss")

vit_valid = np.insert(vit_valid, 0, 2.600)
cnn_valid = np.insert(cnn_valid, 0, 2.1934)

# cnn_train = np.insert(cnn_train, 0, )
# cnn_valid = np.insert(cnn_valid, 0,)
minLength += 1

ax.plot(range(len(cnn_valid[:minLength])), cnn_valid[:minLength], label="CNN-Validation", color = '#49d8fc')
ax.plot(range(len(cnn_train[:minLength])), cnn_train[:minLength], label="CNN-Training", linestyle = 'dotted', color = '#49d8fc',)

ax.plot(range(len(vit_valid[:minLength])), vit_valid[:minLength], label="ViT-Validation", color = '#fcb249')
ax.plot(range(len(vit_train[:minLength])), vit_train[:minLength], label="ViT-Training", linestyle = 'dotted', color = '#fcb249')
ax.set_ylim((0,3))
ax.legend()
plt.title("Pretrained Data Models CE vs. Epochs")
```
